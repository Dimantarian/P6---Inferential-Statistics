## Background
This project will evaluate whether or not the Stroop task highlights a statistically significant difference in the way in which the human brain processes visuals. Data has been collected from a series of trials where the participantâ€™s task is to say out loud the color of the ink in which the word is printed. 

The task has two conditions: 
* A congruent words condition, and an incongruent words condition. In the congruent words condition, the words being displayed are color words whose names match the colors in which they are printed: for example <span style="color:red">**RED**</span>, <span style="color:blue">**BLUE**</span>. 
* In the incongruent words condition, the words displayed are color words whose names do not match the colors in which they are printed: for example <span style="color:orange">**PURPLE**</span>, <span style="color:purple">**ORANGE**</span>. 

In each case, we measure the time it takes to name the ink colors in equally-sized lists. Each participant will go through and record a time from each condition.

## Questions for Investigation
In this section we will set up the statistical analysis of the experiment, detailing the design choices and reasoning.

### Variables
Our **dependent** Variable of for analysis is the time taken to complete the Stroop test. Our **independent variable** or our treatment is whether or not the word, colour pairings are congruent, as described in the background section.

### Hypothesis test structure
The data provided for this test contains two samples, taken across one set of individuals.Given it is a sample of individuals pre/post treatment, and we do not have the descriptive statistics for the whole population (7+bn!), so a **dependant t-Test** is most appropriate. We will be testing the following hypothesis:


<center>**H<sub>0</sub>:** $\mu_0 - \mu_A = 0$ </center>


<center>**H<sub>A</sub>:** $\mu_0 - \mu_A \neq 0$ </center>


In plain english, our *null hypothesis*, *H<sub>A</sub>* is that there is no change in mean time to read the words between pre/post treatment. Our *alternative hypothesis*, *H<sub>A</sub>* is that there is a change.

This requires a two-tailed test.
